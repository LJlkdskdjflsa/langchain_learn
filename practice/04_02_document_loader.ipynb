```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("以太坊白皮书 | ethereum.pdf")
```

```python
documents = loader.load()
```

```python
len(documents)
# 默認一個 doc 就是一頁
```

## 文本轉換器之分割器：RecursiveCharacterTextSplitter

通用型的文本分割器

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(
    # 設置一個非常小的文字塊
    chunk_size=100,
    chunk_overlap=20,
    length_function=len,
)
```

```python
pages = loader.load_and_split(text_splitter=text_splitter)
```

```python
len(pages)
```

## 文本詞嵌入

需要下載模塊： tiktoken (幫忙分片)

```python
from langchain.embeddings import OpenAIEmbeddings

embaddings_model = OpenAIEmbeddings()
```

```python
embeddings = embaddings_model.embed_documents(pages[4].page_content)
```

```python
pages[4].page_content
```

```python
len(embeddings[0])  # 詞向量的維度
```

向量存儲： ChromaDB (本地向量存儲解決方案)

```python
from langchain.embeddings import OpenAIEmbeddings

from langchain.vectorstores import Chroma
```

```python
db = Chroma.from_documents(pages, embaddings_model)
```

相似度檢索

```python
query = "以太坊"

docs = db.similarity_search(query)
```
